### Время аппрува по странам и продуктам

```python
approved_df = median_time_by_countries_and_products.loc[median_time_by_countries_and_products['status_group'] == 'approved', :]
approved_df
```

```python
lifehack = approved_df['country_id'] + ', ' + approved_df['product_id']
```

```python
data=approved_df['conversation_time']
keys=lifehack
palette_color = sns.color_palette('colorblind')
plt.pie(data, labels=keys, colors=palette_color, autopct='%.0f%%')
plt.title('Соотношение времени разговора по странам и продуктам до аппрува')
plt.show()
```

### Время аппрува отдельно по странам и по продуктам

#### По странам

```python
country_pie = df.groupby(['country_id', 'status_group'], as_index=False) \
            .agg({'conversation_time': 'median'})
country_pie_approved = country_pie.loc[country_pie['status_group'] == 'approved', :]
country_pie_approved
```

```python
country_pie_reject = country_pie.loc[country_pie['status_group'] == 'reject', :]
country_pie_reject
```

```python
data=country_pie_approved['conversation_time']
keys=country_pie_approved['country_id']
palette_color = sns.color_palette('colorblind')
plt.pie(data, labels=keys, colors=palette_color, autopct='%.0f%%')
plt.title('Соотношение времени разговора по странам до аппрува')
plt.show()
data=country_pie_reject['conversation_time']
keys=country_pie_reject['country_id']
palette_color = sns.color_palette('colorblind')
plt.pie(data, labels=keys, colors=palette_color, autopct='%.0f%%')
plt.title('Соотношение времени разговора по странам до отказа')
plt.show()
```

#### По продуктам

```python
product_pie = df.groupby(['product_id', 'status_group'], as_index=False) \
            .agg({'conversation_time': 'median'})
product_pie_approved = product_pie.loc[product_pie['status_group'] == 'approved', :]
product_pie_approved
```

```python
product_pie_reject = product_pie.loc[product_pie['status_group'] == 'reject', :]
product_pie_reject
```

```python
data=product_pie_approved['conversation_time']
keys=product_pie_approved['product_id']
palette_color = sns.color_palette('colorblind')
plt.pie(data, labels=keys, colors=palette_color, autopct='%.0f%%')
plt.title('Соотношение времени разговора по продуктам до аппрува')
plt.show()
data=product_pie_reject['conversation_time']
keys=product_pie_reject['product_id']
palette_color = sns.color_palette('colorblind')
plt.pie(data, labels=keys, colors=palette_color, autopct='%.0f%%')
plt.title('Соотношение времени разговора по продуктам до отказа')
plt.show()
```
